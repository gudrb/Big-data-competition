```python
import pandas as pd
import csv
```

```python
enterprise2 = pd.read_csv("기업조사0.csv", engine='python')
```

```python
enterprise2
```

```python
enterprise = pd.read_csv("퇴사율.csv", engine='python')
```

```python
enterprise
```

```python
f_company = pd.merge(enterprise,enterprise2, on = ['name'])
```

```python
f_company
```

```python
import ast
f_company.promotion = f_company.promotion.apply(ast.literal_eval) #숫자로 바꾸기
```

```python
f_company['unpromotion'] = 5.0 - f_company['promotion']
```

```python
f_company = f_company.sort_values(["score"],ascending=[False]) #스코어 내림차순 정렬
```

```python
f_company = f_company.drop(f_company[f_company.r_rate >= 100].index)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(f_company["score"],f_company["r_rate"])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(f_company["순위"],f_company["average"])
```
